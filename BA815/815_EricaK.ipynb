```python
# Load the dataset (update the path if your folder name is different)
import pandas as pd

folder_name = 'BA 815'
csv_file_name = 'Students Social Media Addiction.csv'
csv_path = f'/content/drive/MyDrive/BA 815/Students Social Media Addiction.csv'

df = pd.read_csv(csv_path)
```

Tree map: Distribution of users by academic level and platform. Which Academic level is majorly the target?(Demographic)


```python
import plotly.express as px

# Count students by Academic Level and Platform
count_data = df.groupby(['Academic_Level', 'Most_Used_Platform'], as_index=False).size()

fig = px.sunburst(count_data,
                  path=['Academic_Level', 'Most_Used_Platform'],
                  values='size',
                  color='size',
                  color_continuous_scale='OrRd',
                  title='Treemap: Student Distribution by Academic Level and Most Used Platform<br><sup>N = 705 students</sup>',
                  branchvalues='total')

fig.update_traces(textinfo='label+value+percent parent')
fig.update_layout(margin=dict(t=80, l=10, r=10, b=10),
                  width=1000,
                  height=750,
                  font=dict(size=14))
fig.show()
```

```python
# Cell 2: Quick demographic summary + answer
total_by_level = df['Academic_Level'].value_counts().sort_values(ascending=False)

print("=== Demographic Target Summary ===")
print(total_by_level)
print("\n→ Major target demographic:")
print(f"{total_by_level.index[0]} → {total_by_level.iloc[0]} students ({total_by_level.iloc[0]/len(df)*100:.1f}%)")

# Bonus: biggest single segment
top_segment = count_data.loc[count_data['size'].idxmax()]
print(f"\n→ Dominant single segment:")
print(f"{top_segment['Academic_Level']} + {top_segment['Most_Used_Platform']} → {top_segment['size']} students")
```

```python
# Real rectangular Treemap (네모 버전)
fig_rect = px.treemap(count_data,
                      path=['Academic_Level', 'Most_Used_Platform'],
                      values='size',
                      color='size',
                      color_continuous_scale='Reds',
                      title='Real Rectangular Treemap: Academic Level × Platform')

fig_rect.update_traces(textinfo='label+value+percent parent')
fig_rect.update_layout(margin=dict(t=50, l=0, r=0, b=0))
fig_rect.show()
```

```python
import plotly.express as px

count_data = df.groupby(['Academic_Level', 'Most_Used_Platform'], as_index=False).size()


custom_color_scale = [
    [0.0,   '#FF8C00'],
    [0.3,   '#FFB84D'],
    [0.6,   '#D3D3D3'],
    [1.0,   '#808080']
]

fig = px.treemap(count_data,
                 path=['Academic_Level', 'Most_Used_Platform'],
                 values='size',
                 color='size',
                 color_continuous_scale=custom_color_scale,
                 title='Treemap: Student Distribution by Academic Level and Most Used Platform<br><sup>N = 705 students</sup>')

fig.update_traces(textinfo='label+value+percent parent',
                  hovertemplate='<b>%{label}</b><br>Count: %{value}<br>% of parent: %{percentParent:.1%}')

fig.update_layout(margin=dict(t=80, l=10, r=10, b=10),
                  width=1000,
                  height=650,
                  title_x=0.5,
                  font=dict(size=13),
                  paper_bgcolor='white')

fig.show()
```
